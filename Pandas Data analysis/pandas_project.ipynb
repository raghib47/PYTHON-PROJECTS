### 1. min,max and avg typical hour ?

```python
print('Max Typicaal hours are:', pay['Typical Hours'].max())
print('Min Typicaal hours are:', pay['Typical Hours'].min())
print('Avg Typicaal hours are:', pay['Typical Hours'].mean())
```

### 2. How many employ work on salary and how many employ work on hourly basis?

### 5. What are the max,min and avg ?

```python
pay['Salary'] = pay['Annual Salary'].str.replace('$','')
pay['Salary'] = pay['Salary'].str.replace(',','').astype(float)
```

```python
print('Max Salary:', pay['Salary'].max())
print('Min Salary:', pay['Salary'].min())
print('Avg Salary:', pay['Salary'].mean())
```

### 6. Find the employee who has maximum salary ?

```python
pay[pay['Salary']==pay['Salary'].max()][['Name','Salary']]
```

### 6. Find the employee who has minimum salary ?

```python
pay[pay['Salary']==pay['Salary'].min()][['Name','Salary']]
```

### 7. What are the min,max and avg hourly rate ?

```python
pay['H_Rate']=pay['Hourly Rate'].str.replace('$','').astype(float)
print('min H_Rate', pay ['H_Rate'].min())
print('max H_Rate', pay ['H_Rate'].max())
print('mean H_Rate', pay['H_Rate'].mean().round(2))
```

### 8. How many employee are getting max hourly rate ?

```python
pay[pay['H_Rate']==pay['H_Rate'].max()].count()['Name']
```

### 9. Who is getting max hourly rate ?

```python
pay[pay['H_Rate']==pay['H_Rate'].max()]['Name']
```

### 10. How many employee is getting less than avg hourly rate ?

```python
pay[pay['H_Rate']<pay['H_Rate'].mean()].count()['Name']
```

### 11. How many employees are paid hourly and they have full-time job?

```python
pay[(pay['Salary or Hourly']=='Hourly') & (pay['Full or Part-Time']=='F')].count().head(1)
```

### 12. Find the full-time employees who are working at less than hourly rate of $10.00?
